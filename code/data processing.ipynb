```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
```

```python
df1_3 = pd.read_csv('channel 1.csv')
# rename columns
df1_3 = df1_3.rename(columns={'created_at':'time',
                      'field1':'T1*',
                      'field2':'T1',
                      'field3':'RH1',
                      'field4':'VPD1',
                      'field5':'T3*',
                      'field6':'T3',
                      'field7':'RH3',
                      'field8':'VPD3',
                     })
# set timestamp as index
df1_3['time']= df1_3['time'].str.slice( 0,-6)
df1_3['time'] = pd.to_datetime(df1_3['time'])
df1_3 = df1_3.set_index('time')
df1_3=df1_3[['T1*','T1','RH1','VPD1','T3*','T3','RH3','VPD3']]
df1_3
```

```python
df2_4=pd.read_csv('channel 2.csv')
df2_4=df2_4[['created_at',
             'field1',
             'field2',
             'field3',
             'field4',
             'field5',
             'field6',
             'field7',
             'field8']]
df2_4=df2_4.rename(columns={'created_at':'time',
                      'field1':'T2*',
                      'field2':'T2',
                      'field3':'RH2',
                      'field4':'VPD2',
                      'field5':'T4*',
                      'field6':'T4',
                      'field7':'RH4',
                      'field8':'VPD4',
                     })
df2_4['time']= df2_4['time'].str.slice( 0,-6)

df2_4['time'] = pd.to_datetime(df2_4['time'])
df2_4=df2_4.set_index('time')
```

```python
df5_6=pd.read_csv('channel 3.csv')
df5_6=df5_6[['created_at',
             'field1',
             'field2',
             'field3',
             'field4',
             'field5',
             'field6',
             'field7',
             'field8']]
df5_6=df5_6.rename(columns={'created_at':'time',
                      'field1':'T5*',
                      'field2':'T5',
                      'field3':'RH5',
                      'field4':'VPD5',
                      'field5':'T6*',
                      'field6':'T6',
                      'field7':'RH6',
                      'field8':'VPD6',
                     })
df5_6['time']= df5_6['time'].str.slice( 0,-6)

df5_6['time'] = pd.to_datetime(df5_6['time'])
df5_6=df5_6.set_index('time')
```

```python
df7=pd.read_csv('channel 4.csv')
df7=df7[['created_at',
             'field1',
             'field2',
             'field3',
             'field4'
                    ]]
df7=df7.rename(columns={'created_at':'time',
                      'field1':'T7*',
                      'field2':'T7',
                      'field3':'RH7',
                      'field4':'VPD7',
                      })
df7['time']= df7['time'].str.slice( 0,-6)

df7['time'] = pd.to_datetime(df7['time'])
df7=df7.set_index('time')
```

```python
df=df1_3.join(df2_4).join(df5_6).join(df7)
df1=df.resample('30s').mean()
df1=df1.dropna()
df1.to_csv('all data.csv')
df1
```
