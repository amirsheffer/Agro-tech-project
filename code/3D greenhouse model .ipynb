```python
%matplotlib notebook
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
fig = plt.figure(figsize=(10,5))
ax = Axes3D(fig)

#walls (read as a matrix, every colums is a position vector for the polygon corner)
x =  [0,0,0,  0],      [15,15,15,15],[15,0,0,15],  [15,0,0,15]
y = [0,0,6.5,6.5]  , [0,6.5,6.5,0],[0,0,0,0],    [6.5,6.5,6.5,6.5]
z =  [0,  2.1,2.1,0], [0,0,2.1,2.1],[0,0,2.1,2.1],[0,0,2.1,2.1]

surfaces = []

for i in range(len(x)):
    surfaces.append( [list(zip(x[i],y[i],z[i]))] )

for surface in surfaces:
    ax.add_collection3d(Poly3DCollection(surface,ec='b',fc=(0, 0, 0, 0)))

#roof
x = [15,0,0,15],     [0,15,15,0],          [0,0,0],[15,15,15]
y = [0,0,3.25, 3.25],[6.5,6.5,3.25,3.25],  [0,3.25,6.5],[0,3.25,6.5]
z = [2.1,2.1,3,3],   [2.1,2.1,3,3],        [2.1,3,2.1],[2.1,3,2.1]

surfaces = []

for i in range(len(x)):
    surfaces.append( [list(zip(x[i],y[i],z[i]))] )

for surface in surfaces:
    ax.add_collection3d(Poly3DCollection(surface,ec='r',fc=(0., 0., 0., 0.)))

    
#sensors    
    
    
ax.scatter(xs=[0,7.5,15,0,7.5,15,7.5],
           ys=[-0.05,-0.05,-0.05,6.55,6.55,6.55,3.25],
           zs=[1.7,1.7,1.7,1.7,1.7,1.7,3.05],
           s=40,color='k',
           label='esp32, T and RH',
           marker='d')

ax.scatter(xs=[0,7.5,15,0,7.5,15,12],
           ys=[-0.05,-0.05,-0.05,6.55,6.55,6.55,3.25],
           zs=[0,0,0,0,0,0,3.05],
           s=40,color='y',
           label='T',
           marker='d')

ax.scatter(xs=[0.1,7.5],
           ys=[0.1,6.5],
           zs=[1.5,1.2],
           s=40,color='g',
           label='power supply',
           marker='^')



    
    

ax.set_xlim(0,16)    
ax.set_ylim(0,8)    
ax.set_zlim(0,4)    

fig.suptitle('greenhouse 3D model')
ax.set_xlabel('x (m)')
ax.set_ylabel('y (m)')
ax.set_zlabel('z(m)')
ax.legend(loc=(-0.4,0.1))
plt.rcParams['grid.color'] = (0.5, 0.5, 0.5, 0)
# plt.axis('off')
plt.show()
```
